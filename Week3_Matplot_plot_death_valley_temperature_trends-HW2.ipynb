```python
import numpy as np
from matplotlib import pyplot as plt
import plotly as pt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Reben80/Data110-32213/main/dataset/tempnormals.csv")
```

```python
ds=df[df['location']=='San Diego']
```

```python
descriptive_data_ds = ds['temperature'].describe()
Average_ds = descriptive_data_ds['mean']
```

```python
plt.figure(figsize=(8,5))
plt.style.use('grayscale')

plt.axhline(y=Average_ds, linestyle=':', color='gray',label='Average')
plt.plot(ds['day_of_year'], ds['temperature'], label='San Diego Temperature')
plt.xlabel('Day', labelpad=24)
plt.ylabel('Temperature', labelpad=24)
plt.title('Trend Temperature: San Diego, California',color='black',y=1.1)
plt.grid(True,color='lightgrey', axis='y', )
plt.legend(facecolor='darkgray', framealpha=.8)
plt.show()

```
