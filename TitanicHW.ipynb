```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-32213/main/dataset/titanic.csv')
titanic_df.head()
```

```python
grouped_data = titanic_df.groupby(['class', 'sex']).size().unstack()
grouped_data.plot(kind='bar')

plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class and Sex')
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(titanic_df['class'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(titanic_df, x='age', hue='class', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
plt.hist(titanic_df['age'])

plt.show()
```

```python
plt.hist(titanic_df['age'],bins=5)
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.hist(titanic_df['age'], bins=15, color='lightblue',edgecolor='white')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age(years)')
plt.ylabel('count')
plt.grid(axis='x')
plt.xticks([0, 20, 40, 60])
plt.tick_params(axis='x', colors='gray')  # Set x-axis tick colors to gray
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['age'], color="deeppink", alpha=0.7)
plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['age'], bw_adjust=0.1,  color="green", alpha=1)
plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['age'], bw_adjust=0.4, fill=True, color="purple", alpha=1)
plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['age'], bw_adjust=0.7, fill=True, color="lightblue", alpha=1)
plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['age'], bw_adjust=2, fill=True, color="pink", alpha=1)
plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
ages_male = titanic_df[titanic_df['sex'] == 'male']['age']
ages_female = titanic_df[titanic_df['sex'] == 'female']['age']

bins = 15

plt.figure(figsize=(10, 6))
plt.hist([ages_male, ages_female], bins, label=['Male', 'Female'])

plt.title('Age Distribution by Sex on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(
     [ages_male, ages_female],bins=20, stacked=True,label=['Male', 'Female'],edgecolor='white')


plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Titanic Passengers by Sex')
plt.legend()
plt.grid(axis='x')
plt.xticks([0, 20, 40, 60])
plt.tick_params(axis='x', colors='gray')
```

```python
plt.figure(figsize=(10, 6))


sns.kdeplot(titanic_df['age'], bw_adjust=0.7,fill=True,color="lightgreen", alpha=1, label='Total')

plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))


sns.kdeplot(titanic_df['age'], bw_adjust=0.7, color="black", alpha=1, label='Total')


sns.kdeplot(titanic_df[titanic_df['sex'] == 'male']['age'], bw_adjust=0.7, fill=True, color="orange", alpha=0.7, label='female')

plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.kdeplot(titanic_df['age'], bw_adjust=0.7, fill=True, color="black", alpha=1, label='Total')

sns.kdeplot(titanic_df[titanic_df['sex'] == 'male']['age'], bw_adjust=0.7, fill=True, color="pink", alpha=0.7, label='male')


sns.kdeplot(titanic_df[titanic_df['sex'] == 'female']['age'], bw_adjust=0.7, fill=True, color="lightblue", alpha=0.7, label='female')

plt.title('Kernel Density Estimate of Age for Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
sex_counts = titanic_df['sex'].value_counts()
labels = sex_counts.index
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=labels,)
plt.title('Distribution of Passengers by Sex')

plt.legend()
plt.show()
```

Conclusion

The analysis of survival rates on the Titanic revealed some key factors that significantly influenced the likelihood of survival. Women had a much higher survival rate compared to men, with almost two-thirds of the survivors being female despite making up less than half of the total passengers. Additionally, passengers in first-class cabins were more likely to survive than those in second and third class. Age also played a role, as young children had a high survival rate while young adults in their twenties had a lower chance of survival. These findings highlight the importance of factors such as gender, class, and age in determining survival outcomes during the Titanic tragedy.
