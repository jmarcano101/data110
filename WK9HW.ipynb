```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

DataFrame **Creation**

```python
# Create dataframe
df = pd.DataFrame(
    {
        'Fruit' : ['Apples', 'Oranges', 'Strawberries', 'Blueberries', 'Bananas'],
        'Color': ['green', 'orange', 'red', 'blue', 'yellow'],
        'Count' : [71, 18, 42, 31, 56]
})
```

# Pie Chart

```python
plt.pie(df['Count'], labels=df['Fruit'], autopct='%1.1f%%', colors=df['Color'])
```

Donut Chart

```python
plt.pie(df['Count'], labels=df['Fruit'], autopct='%1.1f%%', colors=df['Color'])
center_circle = plt.Circle((0,0), 0.6, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.show()
```

In general, the donut graph is aesthetically preferable to the pie chart due to its greater use of white space and fewer visual elements to analyze. Additionally, the smaller proportional sizes of each wedge make the differences in size more distinct and easier to distinguish. The only downside is the placement of the percentage labels, which now straddle the inner edge of each segment instead of being neatly inside them, but this can be easily fixed.

Stacked Bar Graph

```python
# Use ggplot style
plt.style.use('ggplot')

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Cumulative sum to stack the bars
cumulative_counts = np.cumsum(df['Count'])

# Create a bar for each party
for i, (fruit, count, color) in enumerate(zip(df['Fruit'], df['Count'], df['Color'])):
    if i == 0:
        ax.bar('Fruit', count, label=fruit, color=color)
    else:
        ax.bar('Fruit', count, bottom=cumulative_counts[i-1], label=fruit, color=color)

# Adding labels
ax.set_ylabel('Count')
ax.legend()

# Show the plot
plt.show()
```

This chart makes it  simpler to compare the size difference between segments, as you can easily compare the height of each section. I may consider adjusting it so that the color labels are included within the bars themselves, rather than in a separate legend, as I find the legend adds unnecessary clutter.
